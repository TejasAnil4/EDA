```python
import pandas as pd

df = pd.read_csv("datasetEDA.csv")
print(df.head())
```

```python
print(df.shape)  # Dimensions of the dataset
print(df.info())  # Data types and missing values
print(df.describe())  # Summary statistics
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
print(df['gender'].unique())
print(df['married'].unique())
# Check unique values in categorical columns
```

```python
df.dropna(inplace=True)  # Remove rows with missing values
#handling missing values
```

```python
df['gender'] = df['gender'].map({'male': 0, 'female': 1})
df['married'] = df['married'].map({'yes': 1, 'no': 0})
#converting data to numerical from categorical
```

```python
df.drop(columns=['rownames'], inplace=True)
#dropping column not required
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df.hist(figsize=(10, 6), bins=20)
plt.show()
#histograms for Numerical Features
```

```python
sns.countplot(x='gender', data=df)
plt.show()

sns.countplot(x='married', data=df)
plt.show()
#countplot
```

```python
sns.scatterplot(x=df['articles'], y=df['prestige'])
plt.show()
#bivariate analysis
```

```python
sns.boxplot(x=df['gender'], y=df['prestige'])
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
#multivariate analysis
```
